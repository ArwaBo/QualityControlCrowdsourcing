```python
import csv
from ete3 import Tree
import pandas as pd 
from difflib import SequenceMatcher
```

# # --- reading the complete file into record

```python
skills_df = pd.read_csv('complete_skills_categories_frequencies.csv',
                        names = ['category', 'skill','value'])
skills_df
```

```python
# data cleaning 
def trimAllColumns(df):
    """
    Trim whitespace from ends of each value across all series in dataframe
    """
    trimStrings = lambda x: x.strip() if type(x) is str else x
    return df.applymap(trimStrings)


skills_df = trimAllColumns(skills_df)
```

```python
# extracting the categories
categories = []
for i in range(0, len(skills_df)):
    category = skills_df.loc[i,'category']
    categories.append(category)
# without repetition
category_set = set(categories)
print(category_set)
print(len(category_set))
```

```python
# Creating the Taxonomy Tree --------------------------------------------
# ---------Level Zero ---Root
taxonomy = Tree()
# ---------Level One --- Categories
for category in category_set:
    taxonomy.add_child(name=category)       
```

```python
taxonomy.write(features=[],format=2, outfile="skills_taxonomy_tree.nw")
print(taxonomy.get_ascii(attributes=["name"]))
taxonomy.show()
```

```python
for catNode in taxonomy:
    for i in range(0, len(skills_df)):
        if catNode.name ==  skills_df.loc[i,'category']:
            catNode.add_child(name=skills_df.loc[i,'skill'])        
```

```python
def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()
```
