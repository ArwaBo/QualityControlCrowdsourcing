## ingredients of Equation #4 in \"Icrowd ---- paper

```python
# the sismilarity matrix S
S =  similarity_matrix_df.values
print(S)
```

```python
"""To create the nomalized similarity, we need to create the Diagonal matrix """ 
# create the diagonal matrix
# get S matrix size (float)\n",
n = numpy.sqrt(S.size) 

# make it integer
n = int(n) 
n   
# set array elements to zero
D = numpy.zeros((n,n))
D
```

```python
# Computing the Diagonal matrix D\n",
for i in range(0,n):
    for j in range(0,n):
        D[i][i] = D[i][i] + S[i][j]
D
```

```python
# Computing the Normalized similarity
Sdash = numpy.zeros((n,n))  # declaration of the normalized similarity matrix 
D_tpomh = numpy.zeros((n,n))  # D to the power of minus 1/2"
```

```python
"""----the diagonal matrix to the power of 1/2-----"""
"""------ compute the inverse of D----"""
for i in range(0, n):
    for j in range(0, n):
        D_tpomh[i][i] = 1/numpy.sqrt(D[i][i])   # only diagonal elements"
```

```python
D_tpomh
```

```python
"""-----the normalized Similarity ------"""
Sdash = D_tpomh @ S @ D_tpomh
```

### we want to estimate the worker estimated accuracy in task t2,t5,t9


```python
# difference between the observed vector and the estimated vector
   
""" Equation (4)"""
# for task in Tasks_Graph.nodes():

worker_Est_Acc_df = pd.DataFrame()   

"""--------creating a vector of zeros to compare with if we reach convergence------"""
k = q_df.loc['@AaRTMART'].count()
z = numpy.zeros((k))  # set array elements to zero
for i in range(0, k):
    z[i] = 0.001  # when should we stop

"""--------performing the offline component--------"""    
for task in Tasks_Graph.nodes():   
    print('task at hand is ', task)
    converge = 1
    q_df.loc['@AaRTMART',:] = 0.0 # set all values to zeros 
    q_df.loc['@AaRTMART',task] = 1.0  # set the value of task at hand to 1 
    
    q = q_df.values
    p = q 
    
    while numpy.any(converge >= z):  # until reach convergence
        pOld = p 
        p = ((1/(1+1.0)* p)  @ Sdash) + ((1.0/(1+1.0)) * q) # applying equation number 4 in Icrowd--
        converge = numpy.subtract(p,pOld)
        
    print('Estimated Accuarcy for task ', task,' is---------------\n', p )
    p_df = pd.DataFrame(p,columns=Tasks_Graph.nodes()) # convert numpy to dataframe
    worker_Est_Acc_df = worker_Est_Acc_df.append(p_df, ignore_index = True)
```
