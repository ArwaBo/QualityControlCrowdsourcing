## TASK MODEL 

### 1 . Task difficulty


#### Task difficulty is computed based on the supply and demand concept, by assessing the tasks requirements and the worker profiles. The model exploits the skills required by the task and the skills known by the crowd. The set of skills required by the task and the set of skills popular in the crowd can be directly compared. The task is classified as difficult when the skills required by task are rare in the crowd

 ### compute the frequencies of skills

```python
skills
```

```python
def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()
```

```python
hash_map = {}

for index, row in skills.iterrows():
    for skill in row : 
        if skill is not '':
            # skill Exist?
    
            if skill in hash_map:
    
                hash_map[skill] = hash_map[skill] + 1
            else:
                hash_map[skill] = 1
hash_map
```

### Create a Hash Map (Dictionary)

```python
map = hash_map   
# write it to a text file

with open('C:/Users/Arwa/Desktop/datasets/Ex1/freelancer_skills_frequency.csv', 'w') as file:
    for skill in map:
        print(skill,',', map[skill], file=file)
```
