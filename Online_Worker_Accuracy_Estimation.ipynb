```python
"""------Compute the worker observed accuracy vector------"""

task_df = pd.read_csv('C:/Users/Arwa/Desktop/datasets/Ex2/Jobs_dataset_diff_with_ave_Ex2.csv')
no_worker = 20
no_task = 20
task2wperfs = pd.Series()
for i in range (0, no_task):
    task2wperfs = task2wperfs.append(computeWorkerPerfs(task_df.loc[i], no_worker))
    
worker_estemated_accuracy_vector = pd.Series()
worker_estemated_accuracy_vector = updateOnPerfs(task2wperfs) 


# print("task2wperfs\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\",task2wperfs)
worker_estemated_accuracy_vector
```

```python
from GetWorkerObservedAccuracyWithTaskDifficulty import Get_Worker_Observed_Accuracy_with_task_diff

def computeWorkerPerfs(task, no_worker):
    Dataset_df = pd.read_csv('C:/Users/Arwa/Desktop/datasets/Ex2/jobs_and_freelancers_combined_Ex2.csv')
    task2wperfs = pd.Series()
    wperfs = pd.Series()


    #  Compute `performance` of each worker in assiged tasks 
    for i in range (0, no_worker):
        if task.loc['Freelancer_username'] == Dataset_df.loc[i,'name']:         
            perf = Get_Worker_Observed_Accuracy_with_task_diff(Dataset_df.loc[i],task)
            
            wperfs[Dataset_df.loc[i,'name']] = perf
            task2wperfs[task.loc['TaskID']] = wperfs
   
    
    
    return task2wperfs
    
```

```python
def updateOnPerfs(task2wperfs):
    """ data frame between worker and set of tasks """  
        
    """ data frame between worker and set of tasks assocaited with values """
     
    """Step 1: For each worker, compute the influencers."""  
    DEFAULT_ACC = 0.5
    accuracies = pd.DataFrame() 
    
    worker2updatedTasks = pd.DataFrame()  

    worker2influs = pd.DataFrame()
    
    for task in task2wperfs.keys():
        wperfs = pd.Series()

        wperfs = task2wperfs[task]

        influs = pd.Series()
        for worker in wperfs.keys():

            temp = pd.Series()
            influs = pd.Series()
            perf = wperfs[worker]
            influs[task] = perf
            temp[worker] = influs 
#         worker2influs = worker2influs.append(temp , ignore_index=False)
            worker2influs.loc[worker,task] = influs[task]
#     print("worker2influs", worker2influs.index)
#  Step 2: Update accuracies for each worker"""
    
    for worker in worker2influs.index:
        updatedTasks = [] 
        influs = pd.DataFrame()
        influs = worker2influs.loc[worker]
        influs = pd.Series(influs)
        p = pd.DataFrame()
        counter = 0
        for influ in influs.keys():
            print("influs[influ]",influs[influ],"worker",worker"---------------------" )
#             if not np.isnan(influs[influ]):
            if counter < 3:
                q = influs[influ]   # incremental weight
#                 print("q", q)
                ppv = PPVs[influ] # PPV of influencer
#                 print("ppv", "of task==============", influ, "is", ppv, "\n")
                if ppv.empty:
                    continue # the PPV does not exist 
                acc = 0.0
                for task in ppv.keys():


                    if acc == "nan" :
                        acc = DEFAULT_ACC # worker cannot be permitted to answer this task
                    acc =   q * ppv[task]


                    p.loc[influ,task] = acc 
#                     print("task",task,"worker",worker, "acc", acc,"q", q , "ppv[task]",ppv[task])
                    updatedTasks.append(task)
                counter = counter + 1 
        print("worker", worker, p)
        accuracies[worker] = p.sum(axis = 0 )
#         print("p.sum", p.sum)
            
    return accuracies
```
