# Step 1 : Generate the similarity graph G on T

```python
from GetTaskSimilarityAlgorithm1 import Algorithm1_Get_task_to_task_similarity
task_df = pd.read_csv('C:/Users/Arwa/Desktop/datasets/Ex2/Jobs_dataset_diff_with_ave_Ex2.csv')

 

def computeSimilarityGraph(simThreshold, no_Tasks):
    similarity_matrix_df = pd.DataFrame()

   
    for i in range (0, no_Tasks):
        for j in range (i-1, no_Tasks):
            if j < 0 : j = 0
            Sim = Algorithm1_Get_task_to_task_similarity(task_df.loc[i],task_df.loc[j])
            print("tasks---------------------------sim",Sim )
            similarity_matrix_df.loc[task_df.loc[i,'TaskID'], task_df.loc[j,'TaskID']] = Sim
            similarity_matrix_df.loc[task_df.loc[j,'TaskID'], task_df.loc[i,'TaskID']] = Sim



    # creating the similarity graph from the similarity matrix

    indices= []
    for i in range(0, no_Tasks):
        indices.append(task_df.loc[i,'TaskID']) # using the jobs names as an index

    indices

    # beacuse the graph nodes can not be a list  it has to a hashble value

    Tasks_Graph = nx.Graph()
    for index in indices:
       Tasks_Graph.add_node(index)

    for node in Tasks_Graph:
        for other_node in Tasks_Graph:
            if node != other_node:
                if round(similarity_matrix_df.loc[node,other_node],2) >= simThreshold:
                     Tasks_Graph.add_edge(node,other_node,weight = round(similarity_matrix_df.loc[node,other_node],2)) # add a wieted edge

    print(similarity_matrix_df)
    return Tasks_Graph 

```

# Step 2 : Normalize the similarity graph 

```python
def NormalizeTheGraph(graph):
    import math


    vertexSum = pd.Series()
    for v in graph.nodes(): 
        sum = 0.0
        edges = graph.edges(v)
        for edge in edges:
            weight = graph.get_edge_data(*edge)
            sum = sum + weight['weight']
        vertexSum[v]= sum
        
    

    for e in graph.edges():
        source = e[0]
        sourceSum = vertexSum[source]
        target = e[1]
        targetSum = vertexSum[target]
        weight = graph.get_edge_data(*e)
        weight = weight['weight']
        weight /= math.sqrt(sourceSum) * math.sqrt(targetSum)
        graph[source][target]['weight']= round(weight,2)
    return graph
```

# Step 3: Compute PP-Vector for each vertex

```python
def simToPPVs(Tasks_Graph, miu): 
    PPVs = pd.DataFrame()
    
    for vertex in Tasks_Graph:
        initPpr = pd.Series()

        """ creating the inital value of q """
        for v in Tasks_Graph:
            if v is vertex : 
                initPpr[v]= 1.0    
            else:
                initPpr[v] = 0.0
        seeds = []
        seeds.append(vertex)
        ppr = pd.Series()
        ppr = doRunPPR(seeds, initPpr, Tasks_Graph, miu)
    
        PPVs[vertex] = ppr

    return PPVs
```

```python
def doRunPPR(seeds, initPpr, graph, miu):
    lambdaa = 1 / (1 + miu)
    ppr = pd.Series()
    ppr = initPpr
    maxIter = 1000
    for iter in range(0, maxIter):
        newPpr = pd.Series()
        for v in ppr.keys(): 
            score1 = (1-lambdaa) * initPpr[v]
            score1 = float(score1)
            edges = graph.edges(v)
            score2 = 0.0 
            totalWeight = 0.0
            for edge in edges: 
                neighbor = edge[0]
                if neighbor is v:
                    neighbor = edge[1]
                weight = graph.get_edge_data(*edge)
                neighborScore = ppr[neighbor]
                neighborScore = float(neighborScore)
                score2 = score2 + weight['weight'] * neighborScore
                totalWeight = totalWeight + weight['weight']
            if score2 > 0.0:
                score2 = score2 / totalWeight
            score2 = score2 * lambdaa  
            score = score1 + score2 
            if v in seeds:
                score = initPpr[v]
            newPpr[v] = score
        
        diff = 0.0
        for v in ppr.keys():
            olds = ppr[v]
            news = newPpr[v]
            diff = diff + abs(olds - news)
        
        ppr = newPpr
        
        if (diff <0.000001):
            break
            

    return ppr
        
           
```

```python
def Graph_drawing(Tasks_Graph):
# drwaing -------------------------------------------------------------------------------
    try:
        import matplotlib.pyplot as plt
    except:
        raise

    elarge =[(u,v) for (u,v,d) in Tasks_Graph.edges(data=True) if d['weight'] >= 0.5] # threshold 0.75
    esmall =[(u,v) for (u,v,d) in Tasks_Graph.edges(data=True) if d['weight'] <0.5]

    pos =  nx.random_layout(Tasks_Graph) # positions for all nodes


    # nodes
    nx.draw_networkx_nodes(Tasks_Graph, pos, node_size=400)

    # edges

    nx.draw_networkx_edges(Tasks_Graph,pos,edgelist=elarge,
                        width=1,edge_color='b')
    nx.draw_networkx_edges(Tasks_Graph,pos,edgelist=esmall,
                        width=1,alpha=0.5,edge_color='g',style='dashed')

    # labels

    edge_labels = dict([((u,v,),d['weight']) for u,v,d in Tasks_Graph.edges(data=True)])
    nx.draw_networkx_labels(Tasks_Graph, pos, font_size=10, font_family='sans-serif')

    nx.draw_networkx_edge_labels(Tasks_Graph,pos,edge_labels=edge_labels,font_size=10,font_family='sans-serif')

    plt.axis('off')
    # plt.savefig(\'task_weighted_similarity_graph.png\') # save as png
    plt.show() # display
```
