```python
task_df =  pd.read_csv('C:/Users/Arwa/Desktop/datasets/Ex2/Jobs_dataset_tokenized_Ex2.csv')

similarity_matrix_df = pd.DataFrame()

for i in range (0,10):
     for j in range (0,10):
        Sim = Algorithm1_Get_task_to_task_similarity(task_df.loc[i],task_df.loc[j])
        similarity_matrix_df.loc[task_df.loc[i, 'TaskID'], task_df.loc[j, 'TaskID']] = Sim

```

```python
S =  similarity_matrix_df.values
print(S)
```

```python
n = numpy.sqrt(S.size) # get S matrix size (float)
n = int(n) # make it integer
D = numpy.zeros((n,n)) # set array elements to zero
```

```python
# create the diagonal matrix 
```

```python
# Computing the Diagonal matrix D
for i in range(0,n):
    for j in range(0,n):
        D[i][i] = D[i][i] + S[i][j]
```

```python
D
```

```python
# Computing the Normalized similarity
Sdash = numpy.zeros((n,n))  # declaration
D_tpomh = numpy.zeros((n,n))  # D to the power of minus 1/2
```

```python
"""----the diagonal matrix to the power of -1/2-----"""
"""------ compute the inverse of D"""
for i in range(0, n):
    for j in range(0, n):
        D_tpomh[i][i] = 1/numpy.sqrt(D[i][i])   # only diagonal elements
```

```python
D_tpomh
```

```python
"""-----the normalized Similarity ------"""
Sdash = D_tpomh @ S @ D_tpomh
```
