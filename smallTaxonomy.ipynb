```python
taxonomy =  Tree("small_skills_taxonomy_tree.nw")
```

```python
#find the depth of the tree 
node,  depth =  taxonomy.get_farthest_leaf()
print(depth)
root = taxonomy.get_tree_root()
depth = int(depth)
for i in range (0,depth+1):
    for node in taxonomy.traverse("postorder"):
        j= i 
        j = float(j)
        if node.get_distance(node,root)== j:
            l = j/depth
            l = round(l,1)
            node.add_feature('level_score',  str(l))

```

```python
taxonomy.write(features=[],format=9,format_root_node=True,
               outfile="skills_small_taxonomy_tree_level_score.nw")
```

```python
from ete3 import Tree, TreeStyle, TextFace,AttrFace
for n in taxonomy.traverse():
    n.add_face(TextFace(n.name, fsize = 16), column=0)
 

root = taxonomy.get_tree_root()
root.add_face(TextFace(root.name, fsize = 16), column=0)
```
