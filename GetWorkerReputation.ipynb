```python
def trimAllColumns(df):
    """
    Trim whitespace from ends of each value across all series in dataframe
    """
    trimStrings = lambda x: x.strip() if type(x) is str else x
    return df.applymap(trimStrings)


combined_df = trimAllColumns(combined_df)
worker_df = trimAllColumns(worker_df)
```

## Extract work history to take the categories of performed tasks

```python
worker_hist = []
for i in range (0, len(combined_df)):
    if combined_df.loc[i,'name'] == worker_df.loc[i,'name']:
        worker_hist.append(combined_df.loc[i])
        
print(worker_hist[0])
```

```python
for i in range(0,len(worker_df)):
        worker_hist = []
        Tcat =  0 
        sum_rates =    worker_df.loc[i,'job_completion_rate'] + \
                       worker_df.loc[i,'on_budget_rate'] + \
                       worker_df.loc[i,'on_time_rate'] + \
                       worker_df.loc[i,'repeat_hire_rate']
        
   
        for j in range (0, len(task_df)):
            if task_df.loc[j,'Freelancer_username'] == worker_df.loc[i,'name']:
                worker_hist.append(task_df.loc[j,'Job_Category'])
                
        for k in range(0, len(worker_hist)):
            Tcat = Tcat + cat_no_task[worker_hist[k]] 
        
        Tw = worker_no_task[worker_df.loc[i,'name']]
        if Tcat == 0:
            Tcat =1
            
        worker_df.loc[i,'sum_rates'] = sum_rates
        worker_df.loc[i,'no_t_cat'] = Tcat
        worker_df.loc[i,'no_tasks'] = Tw
        worker_df.loc[i,'rep'] = sum_rates *(Tw/Tcat)
```
